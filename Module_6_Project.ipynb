### The LastEditedWhen column only has data in 6 of the 230548 rows. This data will not be helpful and my recommendation is that it be dropped from the dataset.

```python
# I would recommed droping the LastEditedWhen column
df.drop(columns=['LastEditedWhen'], inplace = True)
```

### With the LastEditedWhen column removed, the columns that now has the most null values is the TaxAmount column.

```python
df.info()
```

### I observed that StockItemID is a single number, and each number has a corresponding Description. Thus, by knowing the StockItemID of a row, its Description can be imputed. However, if both the StockItemID and Description are missing, the item is unknown.

### This code shows that there are 461 rows that have both a null StockItemID and a null Description?

```python
null_rows = df[df[['StockItemID', 'Description']].isnull().all(axis=1)]
```

```python
null_rows.shape[0]
```

### Since there are only 461 rows and it is not practical to impute the data my recommendation is to drop the rows with null values in the Description and StockItemID columns

```python
df.dropna(subset=['StockItemID', 'Description'], how='all', inplace = True)
```

### After removing those rows we are left with 230087 rows.

```python
df.info()
```

### I noticed there are negative values in the TaxRate column.

```python
df.describe()
```

```python
df['TaxRate'].value_counts()
```

```python
df['TaxRate'].unique()
```

### Based on your domain knowledge and knowing that tax rates should not be a negative number my recommendation is to turn the negative tax rates into positive tax rates by multiplying the column by -1.

```python
df.loc[df['TaxRate'] < 0, 'TaxRate'] *= -1
```

### Now we can see that the negative values have been removed and average tax rate has gone from 14.46 to 14.97.

```python
df.describe()
```

```python
df.info()
```

### I observed that the column TaxAmount also has many null values. This column indicates the amount of tax paid for each invoice line. The TaxAmount can be calculated using the following formula:

#TaxAmount = UnitPrice * Quantity * TaxRate / 100

### Based on this knowledge, my ecommendation would be to impute the missing values

```python
df['TaxAmount'] = df['TaxAmount'].fillna(df['UnitPrice'] * df['Quantity'] * df['TaxRate'] / 100)
```

### After imputing the data, there are no null values and the average TaxAmount hase gone from 113.29 to 112.95.

```python
df.describe()
```

```python
df.info()
```

### The ExtendedPrice column, which indicates the total price paid by the customer, also has null values. The ExtendedPrice can be calculated using the following formula:

# ExtendedPrice = (UnitPrice * Quantity) + TaxAmount

### These null values can also be easily imputed.

```python
df['ExtendedPrice'] = df['ExtendedPrice'].fillna((df['UnitPrice'] * df['Quantity']) + df['TaxAmount'])
```

###We can see that rows with null values in the ExtendedPrice column have been updated.

```python
df.info()
```

```python
df.describe()
```

### I also noticed that some of the values in the ExtendedPrice column were values of -999. I applied the fix to those rows also.

```python
df.loc[df['ExtendedPrice'] == -999, 'ExtendedPrice'] = (df['UnitPrice'] * df['Quantity']) + df['TaxAmount']
```

```python
df.describe()
```

### The updated ExtendedPrice average is 867.64, up from 803.72.

### I will use the z-score method to determine how many outliers there are in the Quantity and ExtendedPrice columns. A thereshold of 3 will be used to determine which rows are outliers in these columns. I will get rows that have a z-score greater than or equal to 3.

```python
average_quantity = df['Quantity'].mean()
std_dev_quantity = df['Quantity'].std()
```

```python
df['Z-Score_quantity'] = (df['Quantity'] - average_quantity) / std_dev_quantity
df.head()
```

```python
df.loc[df['Z-Score_quantity'] >= 3]
```

```python
average_extendedprice = df['ExtendedPrice'].mean()
std_dev_extendedprice = df['ExtendedPrice'].std()
```

```python
df['Z-Score_extendedprice'] = (df['ExtendedPrice'] - average_extendedprice) / std_dev_extendedprice
df.head()
```

```python
df.loc[df['Z-Score_extendedprice'] >= 3]
```

### There are 5987 outliers in the Quantity column and 4522 in the ExtendedPrice column. I have made note of the outliers but do not recommend changing them or removing them from the dataset.

```python
df.head()
```

###The data set provided should have a unique ID for each invoice line, but there can be repeated invoice IDs. This is because several unique invoice lines can share the same invoice ID. Given this information, I will check for duplicates using only the InvoiceLineID column.
