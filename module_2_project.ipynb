### About the data

This data set has 19 coulumns and 6847 rows.
I will use this data to answer the following questions.
- How does land size affect pricing?
- How does property age affect pricing?
- How does pricing compare in different regions?
- Which region has properties built in the widest range of time?

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\rossp\OneDrive\Desktop\BTECH\Adv. Python\housing.csv")
df.info()
```

### How does land size affect pricing?
The scatter plot below has the properties colored by type and shows the land size on the x-axis with the price on the y-axis.

```python
sns.scatterplot(x='Landsize', y='Price', data=df, hue='Type')
plt.show()
```

By adding a linear model line, the parterns are more easily identified. While townhouses and units show only a slight increase in price with larger land sizes, houses have a much steeper price increase.

```python
sns.lmplot(x='Landsize', y='Price', data=df, hue='Type')
plt.show()
```

### How does property age affect pricing?

The line chart below indicates that newer properties tend to have lower prices on average compared to older properties.

```python
mean_price_year_df = df.groupby('YearBuilt').agg({'Price': 'mean'})
```

```python
sns.lineplot(x='YearBuilt', y='Price', data=mean_price_year_df)
plt.show()
```

### How does pricing compare in different regions?

This bar chart show that average prices differ quite a bit when looking at the different regions. 
The Southern Metropoitan region has the highest average price.

```python
mean_price_region_df = df.groupby('Regionname').agg({'Price': 'mean'})
```

```python
plt.figure(figsize=(18, 5))
sns.barplot(x='Regionname', y='Price', data=mean_price_region_df)
plt.show()
```

According to the histogram of the property prices below, the prices are right skewed.

```python
sns.histplot(x='Price', data=df)
plt.show()
```

### Which region has properties built in the widest range of time?

This box plot makes it easy to see which regions have the widest range of time for properties being built.

The Northern Metropolitan region has the widest range.

We can also see the the Western Victoria region has had very few properties listed that were built after the year 2000.

```python
sns.boxplot(x='YearBuilt', y='Regionname', data=df, orient='h')
plt.show()
```

### Conclusion

As expected, prices tend be higher when the land size increases and the property is located in prefered regions like the Southern and Eastern Metropolitan regions.

I was surprised to see the prices tended to be lower for newer properties. It may be that newer properties have less land and/or are built in less desirable regions.
